```python
#Simple Linear Regression
#we need data containing  one input column and one output column
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data_url = "https://raw.githubusercontent.com/yasirmehmood-ai/Machine_Learning-Journey/refs/heads/main/SimpleLinearRegression/Salary_Data.csv"

df = pd.read_csv(data_url)
plt.xlabel("YearsExperience", color="red")
plt.ylabel("Salary", color="red")
plt.scatter(df.YearsExperience, df.Salary)
```

```python
lr= linear_model.LinearRegression()
lr.fit(df[['YearsExperience']],df.Salary)
```

```python
lr.predict([[5.5]])
```

```python
lr.score(df[['YearsExperience']], df.Salary)
```

```python
plt.xlabel("YearsExperience", color="red")
plt.ylabel("Salary", color="red")
plt.scatter(df.YearsExperience,df.Salary)
plt.plot(df.YearsExperience, lr.predict(df[['YearsExperience']]), color="red")
```
